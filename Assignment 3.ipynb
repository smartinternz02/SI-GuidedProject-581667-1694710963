## IMPORT THE NECESSARY LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## IMPORT DATASET

```python
dataset=pd.read_csv("Titanic-Dataset.csv")
dataset
```

## Checking for Null Values

```python
dataset.isnull().any()
```

```python
dataset["Age"].fillna(dataset["Age"].mean(),inplace=True)
dataset
```

```python
dataset["Embarked"].fillna(dataset["Embarked"].mode()[0],inplace=True)
dataset
```

```python
dataset["Cabin"].fillna(dataset["Cabin"].mode()[1],inplace=True)
dataset
```

```python
dataset.isnull().sum()
```

## Outlier Detection

```python
sns.boxplot(dataset.Age)
```

```python
q1=dataset.Age.quantile(0.25)
q1
```

```python
q2=dataset.Age.quantile(0.50)
q2
```

```python
q3=dataset.Age.quantile(0.75)
q3
```

```python
IQR=q3-q1
```

```python
upper_limit=q3+1.5*IQR
upper_limit
```

```python
dataset['Age']=np.where(dataset['Age']>upper_limit,30,dataset['Age'])
```

```python
sns.boxplot(dataset.Age)
```

## Splitting Dependent and Independent variables

```python
#datset.iloc[rows,column]
x=dataset.iloc[:,3:13]
y=dataset.iloc[:,1:2]
```

```python
y.head()
```

```python
x.head()
```

## Perform Encoding

Perform Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x["Sex"]=le.fit_transform(x["Sex"])
x["Sex"]
```

```python
x.head()
```

```python
x.Embarked.value_counts()
```

One hot encoding on geography column

```python
Embarked=pd.get_dummies(x["Embarked"],drop_first=True)
Embarked
```

```python
x=pd.concat([x,Embarked],axis=1)
```

```python
x.head()
```

```python
x.drop(["Embarked"],axis=1,inplace=True)
x.head(6)
```

## Splitting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```
